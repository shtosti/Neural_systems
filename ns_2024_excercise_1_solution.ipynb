## Task 1

**Nernst equation**:
<br>$E_K = \frac{R \cdot T}{z \cdot F} \cdot ln(\frac{K_{out}}{K_{in}})$

**Given**:
<br>$R = 8.314J/mol$ <--- gas constant
<br>$T=37$&deg;$C=37+273.15=310.15$ <--- to Kelvin
<br>$z=+1$ <--- K's valence
<br>$F=96485C/mol$ <--- Faraday's constant
<br>$K_{out}=5mM$
<br>$K_{in}=140mM$

**Plugging the values in**:
<br>$E_K= \frac{8.314J/mol \cdot 310.15}{1 \cdot 96485} \cdot ln(\frac{5mM}{140mM})=-0.089V$

**Solution**:
<br>$E_K=-0.089V$

```python
#Task 2 code:
import numpy as np
import matplotlib.pyplot as plt

# Neuron parameters
resting_membrane_potential = -65.0
firing_threshold = -55.0
membrane_time_constant = 10.0
input_current = 12.0

# Simulation parameters
duration = 20.0
time_step = 0.001

# Initialize variables
time_values = np.arange(0, duration, time_step)
membrane_potential_values = np.zeros_like(time_values)
membrane_potential = resting_membrane_potential

# TODO
# Update membrane potential using forward Euler method -> Task 2a

for i in range(len(time_values)):
    membrane_potential += (time_step / membrane_time_constant) * (resting_membrane_potential - membrane_potential + input_current)
    membrane_potential_values[i] = membrane_potential

    # reset if the firing threshold is reached
    if membrane_potential >= firing_threshold:
        membrane_potential = resting_membrane_potential

# Plot results
plt.plot(time_values, membrane_potential_values, color='blue')
plt.xlabel('Time')
plt.ylabel('Membrane Potential')
plt.title('Integrate-and-Fire Neuron Simulation')
plt.grid(True)
plt.show()
```

### b)

As can be seen below, manipulations to the input current lead to the changes in the firing pattern. If the firing threshold is not reached due to the input current being too small to lead to a sufficient membrane depolarization, the spike is not produced, and we only see an ascending curve. If the input current is large enough, we can see a spike train.

Manipulations to the firing threshold yield similar results: if they make it easier to reach the firing threshold, a spike train is produced, whereas a failure to reach the treshold leaves us without a spike.

Clearly, the model is a gross simplification and does not take into account e.g. hyperpolarization.

```python
#Task 3a code:
import numpy as np
import matplotlib.pyplot as plt

def izhikevich_neuron(a, b, c, d, I, dt=0.01, duration=100):
    """
    Simulate the Izhikevich neuron model.

    Parameters:
        a, b, c, d: Parameters of the neuron model.
        I: Input current.
        dt: Time step (ms).
        duration: Duration of simulation (ms).

    Returns:
        v: Array of membrane potentials over time.
    """
    num_steps = int(duration / dt)
    v = np.zeros(num_steps)
    u = np.zeros(num_steps)

    # TODO
    # Task 3a -> implement here

    # Initial values
    v[0] = -65
    u[0] = b * v[0]

    for t in range(1, num_steps):
        dv = (0.04*v[t-1]**2 + 5*v[t-1] + 140 - u[t-1] + I) * dt
        du = a * (b * v[t-1] - u[t-1]) * dt

        # increment
        v[t] = v[t-1] + dv
        u[t] = u[t-1] + du

        # reset if yhe threshold is reached (with the threshold set to 30)
        if v[t] >= 30:
            v[t] = c
            u[t] += d

    return v

#TODO: from Figure 2 in the paper
# Find a,b,c,d for each

parameters = {
    'RS': (0.02,0.2,-65,8),
    'IB': (0.02,0.2,-55,4),
    'CH': (0.02,0.2,-50,2),
    'FS': (0.1,0.2,-65,2),
    'LTS': (0.02,0.25,-65,2),
    'TC': (0.02,0.25,-65,0.05),
    'RZ': (0.1,0.26,-65,2),
}

I = 10
plt.figure(figsize=(8, 8))
for i, (neuron_type, params) in enumerate(parameters.items()):
    a, b, c, d = params
    membrane_potential = izhikevich_neuron(a, b, c, d, I)
    plt.subplot(4, 2, i+1)
    plt.plot(membrane_potential)
    plt.title(neuron_type)
    plt.xlabel('Time (ms)')
    plt.ylabel('Membrane Potential (mV)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

### b)

```python
#Task 3b code:
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 0.02
b = 0.25
c = -65
d = 0.05

# Define Izhikevich neuron model
def izhikevich_neuron(I, v_init=-65, u_init=0, dt=0.01, T=50):
    # Task 3b -> implement using same method as 3a + count spikes
    # Compute frequency from spike count and T

    num_steps = int(T / dt)
    v = np.zeros(num_steps)
    u = np.zeros(num_steps)
    spike_count = 0

    for t in range(num_steps):
        dv = (0.04*v_init + 5) * v_init + 140 - u_init + I
        du = a * (b*v_init - u_init)

        # increment
        v_init += dv * dt
        u_init += du * dt

        # reset once the threshold has been reached
        if v_init >= 30:
            spike_count += 1
            v_init = c
            u_init += d

    # NB divide by 1000 to convert to seconds!
    s = spike_count / (T/1000)

    return s

# Define range of input currents
I_values = np.linspace(0, 100, 100)
firing_rates = [izhikevich_neuron(I) for I in I_values]

# Plot FI curve
plt.plot(I_values, firing_rates)
plt.title('Frequency-Current (FI) Curve')
plt.xlabel('Input Current (I)')
plt.ylabel('Firing Frequency (Hz)')
plt.grid(True)
plt.show()
```

```python
#Task 4a code:
import numpy as np
import matplotlib.pyplot as plt

def voltage_profile_ratio(x, V0, R, Rm, Cm):
    """
    Calculates the voltage profile ratio Vx/V0 along the cable.

    Parameters:
        x (array-like): Array of positions.
        V0 (float): Initial voltage at origin.
        R (float): Axial resistance per unit length.
        Rm (float): Membrane resistance per unit length.
        Cm (float): Membrane capacitance per unit length.

    Returns:
        array: Voltage profile ratio Vx/V0.
    """
    # TODO: Complete function
    # NB: use absolute values
    Vx_V0 = np.exp(- (R + Rm) * np.abs(x) / (Rm * Cm))

    return Vx_V0


# Define parameters
V0 = 1.0  # Initial voltage at origin
R = 1.0  # Axial resistance per unit length
Rm = 1.0  # Membrane resistance per unit length
# NB: setting capacitance to 1.0 to provide the function argument
Cm = 1.0

# Generate positions
x = np.linspace(-5, 5, 100)

# Calculate voltage profile ratio
Vx_V0 = voltage_profile_ratio(x, V0, R, Rm, Cm)

# Plot
plt.plot(x, Vx_V0, label='Voltage Profile Ratio')
plt.xlabel('Position (x)')
plt.ylabel('Voltage Profile Ratio (Vx/V0)')
plt.title('Voltage Profile Ratio along Cable')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
